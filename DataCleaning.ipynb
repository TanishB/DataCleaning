# Data Cleaning
## Tackle the Null Values
# 3 ways to do it
- Drop the Corresponding rows
- Drop that attribute(column)
- Set value to Zero, Mean, Median etc

```python
import pandas as pd
data = pd.read_csv('titanic.csv')
```

```python
data.head()
```

# From above we can see age attribute has Null Values
## 1. Drop the Corresponding rows

```python
temp1 = data
temp1.describe()
```

```python
temp1 = temp1.dropna(subset=['Age'])
temp1.describe()
```

## 2. Drop that attribute(column)

```python
temp2 = data
temp2.head()
```

```python
temp2 = temp2.drop('Age' , axis=1)
temp2.head()
```

## 3. Set value to Zero, Mean, Median etc

```python
temp3 = data
temp4 = data
temp5 = data
```

## Setting Null Values to Zero

```python
temp3['Age'] = temp3['Age'].fillna(0)
temp3.describe()
```

## Setting Null values to median

```python
median = temp4['Age'].median()
temp4['Age'] = temp4['Age'].fillna(median)
temp4.describe()
```

## Setting Null values to mean

```python
mean = temp5['Age'].mean()
temp5['Age'] = temp5['Age'].fillna(mean)
temp5.describe()
```

# Another Handy Way

```python
from sklearn.preprocessing import Imputer
```

```python
imputer  = Imputer(strategy = 'median')
```

```python
x = data
```

```python
x.describe()
```

```python
x.head()
```

# Imputer only works on numerical attributes so lets drop the categorical ones for now

```python
x = x.drop(['Name' , 'Sex' , 'Ticket' , 'Cabin' , 'Embarked'],axis=1)
```

```python
x.head()
```

# -> Fit -> Transform

```python
#First we have to calculate the median
imputer.fit(x)
```

```python
#Then we apply it on the data and store the result in a new name
X = imputer.transform(x)
#Watch it, X is an numpy array, we need to convert it into dataframe
```

```python
X = pd.DataFrame(X, columns=x.columns)
```

```python
X.describe()
#Null values are filled (Check Age count below)
```
